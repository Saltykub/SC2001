```python
random.seed(123)
dataset = []
for i in range(1,8):
    data = generate_sample(10**i)
    dataset.append(data)
```

```python
keyComps = []
for i in range(1,1001):
    keyComps.append(merge_sort_hybrid(dataset[2],0,len(dataset[2])-1,i))
```

```python
# Create a list of S values from 1 to 500
s_values = list(range(1, 1001))

# Create a dataframe for the plot
import pandas as pd
df = pd.DataFrame({
    'S': s_values,
    'Key Comparisons': keyComps
})
# Plotting using plotly.express
fig = px.line(df, x='S', y='Key Comparisons', title='Key Comparisons vs. Threshold Value (S)',
              labels={'S': 'Threshold Value (S)', 'Key Comparisons': 'Number of Key Comparisons'})
fig.show()
```
