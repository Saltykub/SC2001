```python
from utils import *
import time
import plotly.express as px
import pandas as pd
import random 
import numpy as np
```

```python
random.seed(123)
dataset = []
for i in range(1,8):
    data = generate_sample(10**i)
    dataset.append(data)
```

```python
keyComps = []
for i in range(1,1001):
    temp = dataset[2].copy()
    keyComps.append(merge_sort_hybrid(temp,0,len(dataset[2])-1,i))
```

```python
# Create a list of S values from 1 to 500
s_values = list(range(1, 1001))

# Create theoretical  value 
n = 10**3
arr = np.arange(1, 1001)
theory = n*arr + n*(np.log2(n / arr))

# Create a dataframe for the plot
import pandas as pd
df = pd.DataFrame({
    'S': s_values,
    'Key Comparisons': keyComps,
    'Theoretical key Comparisons' : theory
})
# Plotting using plotly.express
fig = px.line(df, x='S', y=['Key Comparisons','Theoretical key Comparisons'], title='Key Comparisons vs. Threshold Value (S)',
              labels={'S': 'Threshold Value (S)', 'Key Comparisons': 'Number of Key Comparisons'})
fig.show()
```
