```python
from utils import *
import pandas as pd
import time
import plotly.express as px
```

```python
random.seed(123)
dataset = []
for i in range(1,8):
    data = generate_sample(10**i)
    dataset.append(data)
```

```python
import time
t = []
keys = []
for s in range (1,301):
    temp = dataset[6].copy()
    t_start = time.time()
    key = merge_sort_hybrid(temp,0,len(dataset[6])-1,s)
    t_stop = time.time()
    keys.append(key)
    t.append(t_stop-t_start)
    print(f"finish lapes:{s}")
```

```python
import numpy as np

n = 10**7
arr = np.arange(1, 1001)
theory = n*arr + n*(np.log2(n / arr))
```

```python
s = list(range(1, 301))
df = pd.DataFrame({"times":t, "keys" : keys, "s":s,"theory":theory})
fig = px.line(df, x="s", y="times", title='Time complexity for each S')
fig.show()
```
